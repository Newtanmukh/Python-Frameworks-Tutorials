```python
pokemon = pl.read_csv('data/pokemon_data.csv')
```

```python
pokemon.head()
```

```python
pokemon.select(pl.col("*"))
```

```python
pokemon.select(pl.col("Name"))
```

```python
pokemon.filter(
    (pl.col("Legendary")==True) &
    (pl.col("Type 1")=="Grass")
).head()
```

```python
print(pokemon.sample(5))
```

```python
pokemon.columns
```

```python
pokemon.select(
    pl.col("Name").sort(),
    pl.col("HP").mean().alias("Average HP"),
    pl.col("Attack").max().alias("Max Attack"),
    pl.col("Attack").min().alias("Min Attack"),
    pl.col("Speed")
).head(7)
```

```python
pokemon.filter(
    pl.col("HP") > 150
)
```

```python
gen1 = pokemon.filter(
    pl.col("Generation")==1
)

out = gen1.group_by("Type 1").agg(
    pl.col("#").count()
)

out.select(
    pl.col("Type 1"),
    pl.col("#").sort(descending = True)
)
```

```python
out.head(2)
```

```python
out[["Type 1", "#"]].head(3)
```

```python
pokemon.filter(
    pl.col("Name").str.starts_with("B")
).head(3)
```
