# SCatter Plot

```python
sns.set(style='whitegrid')
fmri = sns.load_dataset("fmri")
 
sns.scatterplot(x="timepoint",
                    y="signal",
                    hue="region",
                    style="event",
                    data=fmri)
```

# Historam Plot

```python
df = sns.load_dataset('tips')
sns.set_style('whitegrid')
sns.distplot(df['total_bill'], kde = False, color ='red', bins = 100) #Number of rectangular subplots
```

```python
df = sns.load_dataset('tips')
sns.set_style('whitegrid')
sns.distplot(df['total_bill'], hist = False, color ='red', bins = 30,vertical=True)
```
